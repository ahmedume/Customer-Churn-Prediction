# EDA — Telco Customer Churn

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('../data/WA_Fn-UseC_-Telco-Customer-Churn.csv')
df.head()
```

```python
print('Shape:', df.shape)
print('\nDtypes:')
df.dtypes
```

```python
print('Missing values:\n', df.isnull().sum())
print('\nDuplicates:', df.duplicated().sum())
```

```python
print('Target distribution:')
print(df['Churn'].value_counts())
print(df['Churn'].value_counts(normalize=True))
```

```python
df.describe(include='all')
```

```python
fig, axes = plt.subplots(1, 2, figsize=(10, 4))
df['Churn'].value_counts().plot(kind='bar', ax=axes[0], title='Churn Count')
df['Churn'].value_counts().plot(kind='pie', ax=axes[1], title='Churn Proportion', autopct='%.1f%%')
plt.tight_layout()
plt.show()
```

```python
num_cols = df.select_dtypes(include=[np.number]).columns.tolist()
if num_cols:
    df[num_cols].hist(figsize=(12, 8), bins=30)
    plt.tight_layout()
    plt.show()
```

```python
cat_cols = df.select_dtypes(include=['object']).columns.tolist()
cat_cols.remove('customerID')
cat_cols.remove('Churn')

fig, axes = plt.subplots(len(cat_cols)//3 + 1, 3, figsize=(15, 4*(len(cat_cols)//3 + 1)))
axes = axes.flatten()
for i, col in enumerate(cat_cols):
    pd.crosstab(df[col], df['Churn'], normalize='index').plot(kind='bar', stacked=True, ax=axes[i])
    axes[i].set_title(f'{col} vs Churn')
for j in range(i+1, len(axes)):
    axes[j].set_visible(False)
plt.tight_layout()
plt.show()
```
